```python
options = webdriver.ChromeOptions()
path ='/usr/local/bin/chromedriver'
driver_1 = webdriver.Chrome(executable_path=path, options=options)
```

```python
job_data_3 = []
# list = ['Mumbai','Delhi','Bengaluru','Hyderabad']


for page in range(0,70 ,10):

    url =f'https://in.indeed.com/jobs?q=data+analyst&l=Hyderabad&start={page}'
    driver_1.get(url)
    time.sleep(random.uniform(8,11))

    try:
        close = driver_1.find_element(By.XPATH,'//button[@class = "icl-CloseButton icl -Modal-close"]')
        close.click()
    except:
        pass

    jobs = driver_1.find_elements(By.XPATH,'//div[@class ="css-1m4cuuf e37uo190"]')


    for job in jobs:
        job.location_once_scrolled_into_view
        try:
            close = driver_1.find_element(By.XPATH,'//button[@class = "icl-CloseButton icl -Modal-close"]')
            close.click()
        except:
            pass

        job.click()
        time.sleep(random.uniform(5,7))

        try:
            job_title = driver_1.find_element(By.XPATH,'//h2[@class = "icl-u-xs-mb--xs icl-u-xs-mt--none jobsearch-JobInfoHeader-title is-embedded"]').text.strip()
            title = job_title.split('\n')
        except:  
            title ='Nan'

        try:
            company = driver_1.find_element(By.XPATH,'//div[@class = "css-1h46us2 eu4oa1w0"]').text.strip()
            company = company.split('\n')
        except:
            company ='Nan'

        try:
            location = driver_1.find_element(By.XPATH,'//div[@class = "css-6z8o9s eu4oa1w0"]').text.strip()
            location = location.split('\n')
        except:
            location = 'Nan'

        try:
            salary = driver_1.find_element(By.XPATH,'//span[@class = "css-2iqe2o eu4oa1w0"]').text.strip()
            salary = salary.split('/n')
        except:
            salary ='Nan'

        try:
            rating = driver_1.find_element(By.XPATH,'//span[@class = "ratingNumber"]').text.strip()
            rating = rating.split('/n')
        except:
            rating ='Nan'

        try:
            job_type = driver_1.find_element(By.XPATH,'//div[@class = "css-rr5fiy eu4oa1w0"]').text.strip()
            job_type = job_type.split('/n')
        except:
            job_type ='Nan'

        try:
            experience = driver_1.find_element(By.XPATH,'//div[@class = "css-1kvmpfe eu4oa1w0"]').text.strip()
            experience = experience.split('/n')
        except:
            experience ='Nan'

#         time.sleep()

#             job_description = driver_1.find_element(By.XPATH,'//div[@id="jobDescriptionText"]').text.strip()
        try:
            job_description = driver_1.find_element(By.XPATH,'//div[@id="jobDescriptionText"]').text.strip()
        except:
            job_description = 'Nan'
#         time.sleep(5)

        data = {'job_title':title[0], 'company':company[0] , 'location' :location[0], 'salary' :salary[0],
                'rating' : rating[0],
               'job_type' : job_type[0], 'Experience' : experience[0],'job_description' : job_description}

        job_data_3.append(data)
        print('*Saving')

print(len(job_data_1))



        
        
```
