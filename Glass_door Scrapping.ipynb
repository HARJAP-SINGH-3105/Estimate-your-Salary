```python
from selenium import webdriver
import time
from time import sleep
import pandas as pd
from bs4 import BeautifulSoup
from selenium.webdriver.common.by import By
import requests
import random
```

```python
options = webdriver.ChromeOptions()
path ='/usr/local/bin/chromedriver'
driver = webdriver.Chrome(executable_path=path, options=options)
```

```python
job_data= []
```

```python
# Given url has replaced manually while scrapping for different job roles 
url ='https://www.glassdoor.co.in/Job/bangalore-software-jobs-SRCH_IL.0,9_IC2940587_KO10,18_IP2.htm?includeNoSalaryJobs=true&pgc=AB4AAYEAHgAAAAAAAAAAAAAAAf%2BFBtcASwEBAQgC%2BFj4HKpph55bcVc6%2FSqU6PpZjIrP%2FZUYaNeMyNd%2FDfRTVltkOm%2B3qr0B5lnH3je7DhQ5ZqX3ht%2FoR2dkyBRGmJnFte1HrAAA'
driver.get(url)
# # 
jobs = driver.find_elements(By.XPATH,'//li[@class ="react-job-listing css-19d5y1n eigr9kq3"]')


for job in jobs:
    

    try:
        close = driver.find_element(By.XPATH,'//span[@class = "SVGInline modal_closeIcon"]')
        close.click()
    except:
        pass        
    job.click()
    time.sleep(random.uniform(5,7))

    try:
        job_title = driver.find_element(By.XPATH,'//div[@class = "css-1vg6q84 e1tk4kwz4"]').text.strip()
        title = job_title.split('\n')
    except:  
        title ='Nan'     
    print(title)


    try:
        rating = driver.find_element(By.XPATH,'//span[@class = "css-1m5m32b e1tk4kwz2"]').text.strip()
        rating = rating.split('/n')
    except:
        rating ='Nan'   
    print(rating)


    try:
        salary = driver.find_element(By.XPATH,'//div[@class = "css-1bluz6i e2u4hf13"]').text.strip()
#         salary = driver.find_element(By.XPATH,'//div[@class = "css-3g3psg pr-xxsm"]').text.strip()
        salary = salary.split('/n')
    except:
        salary ='Nan'
    print(salary)


    try:
        location = driver.find_element(By.XPATH,'//div[@class = "css-56kyx5 e1tk4kwz5"]').text.strip()
        location = location.split('\n')
    except:
        location = 'Nan'
    print(location)


    try:
        company = driver.find_element(By.XPATH,'//div[@class = "css-87uc0g e1tk4kwz1"]').text.strip()
        company = company.split('\n')
    except:
        company ='Nan'
    print(company)


    try:
        revenue= driver.find_element(By.XPATH,'//div[@class = "d-flex flex-wrap"]').text.strip()
        revenue= revenue.split('\n')
        print(len(revenue))
        rev_dic ={}
        index =0
        for i in range(int(len(revenue)/2)):
            rev_dic[revenue[index]]= revenue[index+1]
            index= index+2
    except:
        rev_dic={}
        rev_dic['Type']='Nan'
        rev_dic['Revenue']= 'Nan'
        
    print(rev_dic['Type'],rev_dic['Revenue'])
    
    data = {'job_title':title[0], 'company':company[0] , 'location' :location[0], 'salary' :salary[0],
                'rating' : rating[0],
               'revenue': rev_dic['Revenue'], 'type':rev_dic['Type']}
    job_data.append(data)
    job.location_once_scrolled_into_view
   
```

```python
driver.quit()
```
